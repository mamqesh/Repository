```python
#ВАРИАНТ 1
#1 Загрузите данные о выживаемости пассажиров Титаника из файла titanic_train.csv
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("titanic_train.csv", sep = ",")
```

```python
#2 Представьте данные в виде таблицы. Посмотрите на первые 5 строк.
df.head()
```

```python
#3 Выведите на экран основную информацию (info) о наборе данных и признаках
df.info()
```

```python
#4 Выведите на экран данные о самом пожилом мужчине, находящемся на борту
df["Sex"] =="male"
df["Age"].max()
```

```python
#5 Выведите распределение переменной Pclass по всем классам 
#только для мужчин / женщин по отдельности
df["Pclass"].value_counts()
```

```python
#5 Сколько было мужчин 2-го класса?
df[(df["Pclass"] == 1 ) & (df["Sex"] == "female")]
```

```python
#6 Сколько мужчин за 50 лет выжили после крушения?
df[(df["Survived"] == 1 ) & (df["Age"] > 49)& (df["Sex"] == "male")]
```

```python
#7 Подсчитать средний возраст женщин погибших при крушении
df[df["Sex"] == "female"] ["Age"].mean()
```

```python
#8 Постройте попарные зависимости признаков Age, Fare, Pclass, Sex, SibSp, Parch, Embarked и Survived. 
#(метод scatter_matrix Pandas или pairplot Seaborn).
sns.set_theme(style="darkgrid")
sns.pairplot(df[["Age", "Fare", "Pclass", "Sex", "SibSp", "Parch", "Embarked", "Survived"]])
```

```python
#9 Постройте гистограммы распределения каждого признака (столбца данных).
df.hist()
plt.show()
```

```python
#10  Каково соотношение погибших и выживших в зависимости от класса каюты? 
#Отобразите c помощью Seaborn.countplot c аргументом hue.
sns.countplot(x="Survived", hue="Pclass", data = df)
plt.show()
```

```python
#11 Как факт выживания зависит от возраста пассажира? Проверьте (графически) предположение, что молодые чаще выживали. 
#Пусть, условно, молодые - младше 30 лет, пожилые – старше 60 лет.

df ['AgeGroup'] = pd.cut (df ['Age'], 4) # Значение столбца age на пять равных частей.
df.AgeGroup.value_counts(sort=False)

plt.figure(figsize=(10,5))
sns.barplot(data=df,x='AgeGroup',y='Survived',ci=None)
plt.xticks (rotation = 60) 
 
plt.show()
```
